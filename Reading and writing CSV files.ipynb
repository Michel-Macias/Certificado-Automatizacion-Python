```python
import csv
import os

os.chdir('/home/michelmacias/Documentos/FORMACION/GOOGLE/Cuadernos1/CuadernosJupyter/csv/cf/')

with open('cf0102.csv', 'r') as csv_file:
# uso de next() para omitir la primera linea
    # lector_csv es una lista
    lector_csv = csv.reader(csv_file)
    next(lector_csv)
    
    # Creamos file csv nuevo
    with open('new_names.csv', 'w') as new:
        escritor_csv = csv.writer(new, delimiter='\t')
        # delimiter para el separador( delimiter = '*')
        
        for line in lector_csv:
            print((line)[0] + ' ' + (line)[2])
            escritor_csv.writerow(line)
            # print(line)   
```

```python
# Usamos delimiter para abrir el file anterior sin delimiter
# y nos devuelve -> [MAR 1995\t3644\t23977\t19542\t9137]
with open('new_names.csv', 'r') as csv_file:
    # pasamos delimiter='\t'
    lector_csv = csv.reader(csv_file, delimiter='\t')
    for line in lector_csv:
        print(line[:4])
        
```

```python
dir(csv.DictReader)
print(help(csv.DictReader))
```

```python
# Tenemos una lista con keys/diccionario -->'name', 'username', departament
# creamos una lista con estas claves , abrimos un csv y lo rellenamos 
# con la info del diccionario anterior
users = [ 
    {'name': 'Sol Nansy', 'username':'solm', 'departament': 'IT Infraestructure'},
    {'name': 'Liam Nelson', 'username':'lion', 'departament': 'UIX'},
    {'name': 'Charly Grey', 'username':'greyc', 'departament': 'DevOps'},
    
]
keys = ['name','username','departament']
with open('por_departamento.csv', 'w') as p_d:
    writer = csv.DictWriter(p_d, fieldnames = keys)
    writer.writeheader()
    writer.writerows(users)
    
# Para ver el file csv recien creado a prtir de los datos pasados
# abrimos el file con open()
with open('por_departamento.csv', 'r') as csv_file:
    lector_csv = csv.DictReader(csv_file)
    indice = 1
    for line in lector_csv:
        # print(line['name'])
        # ---> Sol Nansy
        #      Liam Nelson
        #      Charly Grey
        # cambiando la key del diccionario nos devuelve el dato asociado
        print('Datos del CSV: --->',indice)
        print(f"Nombre: {line['name']} \nUsuario: {line['username']} \nDepartamento: {line['departament']}")
        indice +=1
        print()
```
